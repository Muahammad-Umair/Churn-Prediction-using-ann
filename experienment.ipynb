```python
import pandas as pd
```

```python
churn_pd=pd.read_csv("Churn_Modelling.csv")
churn_pd
```

```python
churn_pd=churn_pd.drop(["RowNumber", "CustomerId", "Surname"],axis=1)
churn_pd
```

Now first we need to convert the gender into binary form

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder= LabelEncoder()
label_encoder_gender=labelEncoder.fit_transform(churn_pd['Gender'])
churn_pd['Gender']=label_encoder_gender
churn_pd
```

Next Thing we need to convert the geography into numerical form.

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder=OneHotEncoder()
one_hot_encoder_geo=one_hot_encoder.fit_transform(churn_pd[["Geography"]]).toarray()
one_hot_encoder_geo
```

```python
feature_names=one_hot_encoder.get_feature_names_out(['Geography'])
feature_names
```

```python
geo_encoder_df=pd.DataFrame(
    one_hot_encoder_geo,
    columns=feature_names
)
geo_encoder_df
```

```python
churn_pd=pd.concat([churn_pd.drop("Geography",axis=1),geo_encoder_df],axis=1)
churn_pd
```

Now next step seprate the the output from the dataframe

```python
features= churn_pd.drop("Exited",axis=1)
features
```

```python
target=churn_pd['Exited']
target
```

Now we need to scalar it so its value between 0 or the 1

```python
from sklearn.preprocessing import StandardScaler
standardScaler= StandardScaler()
features=standardScaler.fit_transform(features)
features
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(features, target,  test_size=0.2, random_state=42)
```

Start saving the file

```python
import pickle
with open("label_encoder_gender.pkl",'wb') as file:
    pickle.dump(labelEncoder,file)
```

```python
with open("one_hot_encoder_geo.pkl",'wb') as file:
    pickle.dump(one_hot_encoder,file)
```

```python
with open("scaler.pkl",'wb') as file:
    pickle.dump(standardScaler, file)
```

Start making the models

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model=Sequential([
    Dense(64,activation="relu", input_shape=(features.shape[1],)),
    Dense(32,activation="relu", ),
    Dense(1,activation='sigmoid',),
]
)
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import TensorBoard
import datetime

log_dir="logs/fits/"+datetime.datetime.now().strftime("%Y%m$d-%H%M%S")
tensorBoard_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)

```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlyStopping=EarlyStopping(
    monitor="val_accuracy",
     patience=10, 
    restore_best_weights=True
)
```

```python
import keras
optimizer=keras.optimizers.Adam(learning_rate=0.02)
loss=keras.losses.BinaryCrossentropy(),
matrics=keras.metrics.BinaryAccuracy(),
```

```python
model.compile(
     optimizer=optimizer,
     loss=loss,
     metrics=["accuracy"]
)
```

```python
model.fit(
x_train,
y_train,
epochs=50,
validation_data=[x_test,y_test],
callbacks=[tensorBoard_callback,earlyStopping]
)
```
